```python
#3 
#use .assign and create the column name and assign it to a list 
existent_coapplicant = [True if val== 0 else False for val in df_encoded['CoapplicantIncome']]
df_encoded = df_encoded.assign(ExistentCoapplicantIncome=existent_coapplicant)
#reorder the columns 
df_encoded = df_encoded[['Loan_ID', 'Married', 'Education', 'Self_Employed',
       'ApplicantIncome', 'CoapplicantIncome', 'ExistentCoapplicantIncome','LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Loan_Status', 'Dependents_0',
       'Dependents_1', 'Dependents_2', 'Dependents_3+', 'Gender_Female',
       'Gender_Male', 'Gender_Other', 'Property_Area_Rural',
       'Property_Area_Semiurban', 'Property_Area_Urban']]
```

```python
#1 
Loan_Income_Ratio = df['LoanAmount']/df['ApplicantIncome']
df_encoded = df_encoded.assign(Loan_Income_Ratio=Loan_Income_Ratio)
```
